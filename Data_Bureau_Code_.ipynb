```python
import pandas as pd

# Loading train and test data
train_data = pd.read_csv('Assignment_Train.csv')
test_data = pd.read_csv('Assignment_Test.csv')

# Loading feature dictionary
feature_dict = pd.read_excel('Assignment_FeatureDictionary.xlsx')
```

Handling Missing values

```python
# Checking for missing values
print(train_data.isnull().sum())
```

```python
#Dropping columns with more than 4500 NaN values
train_data=train_data.drop(['MIDDLE NAME','Cibil Score','TOTAL ASSET COST','ASSET CTG','MARITAL STATUS','EMPLOY CONSTITUTION','EMPLOYER NAME','EMPLOYER TYPE','Phone Social Premium.a23games','Phone Social Premium.instagram','Phone Social Premium.isWABusiness','Phone Social Premium.jiomart','Phone Social Premium.my11','Phone Social Premium.rummycircle','Phone Social Premium.whatsapp','Phone Social Premium.yatra'],axis=1)
```

```python
#Keeping rows with atleast 36 non NaN values
train_data=train_data.dropna(thresh=36)
```

```python
# Same steps for test data
test_data=test_data.drop(['MIDDLE NAME','Cibil Score','TOTAL ASSET COST','ASSET CTG','MARITAL STATUS','EMPLOY CONSTITUTION','EMPLOYER NAME','EMPLOYER TYPE','Phone Social Premium.a23games','Phone Social Premium.instagram','Phone Social Premium.isWABusiness','Phone Social Premium.jiomart','Phone Social Premium.my11','Phone Social Premium.rummycircle','Phone Social Premium.whatsapp','Phone Social Premium.yatra'],axis=1)
```

```python
#Filling the missing values with relevant values (None, 0)
test_data['HDB BRANCH STATE']=test_data['HDB BRANCH STATE'].fillna('None')
test_data['LAST NAME']=test_data['LAST NAME'].fillna('None')
test_data['ADDRESS TYPE']=test_data['ADDRESS TYPE'].fillna('None')
test_data['Pan Name']=test_data['Pan Name'].fillna('None')
test_data['vpa']=test_data['vpa'].fillna('None')
test_data['upi_name']=test_data['upi_name'].fillna('None')
test_data['HDB BRANCH NAME']=test_data['HDB BRANCH NAME'].fillna('None')
test_data['Phone Social Premium.amazon']=test_data['Phone Social Premium.amazon'].fillna(0)
test_data['Phone Social Premium.flipkart']=test_data['Phone Social Premium.flipkart'].fillna(0)
test_data['Phone Social Premium.jeevansaathi']=test_data['Phone Social Premium.jeevansaathi'].fillna(0)
test_data['Phone Social Premium.microsoft']=test_data['Phone Social Premium.microsoft'].fillna(0)
test_data['Phone Social Premium.paytm']=test_data['Phone Social Premium.paytm'].fillna(0)
test_data['Phone Social Premium.shaadi']=test_data['Phone Social Premium.shaadi'].fillna(0)
test_data['Phone Social Premium.skype']=test_data['Phone Social Premium.skype'].fillna(0)
test_data['Phone Social Premium.toi']=test_data['Phone Social Premium.toi'].fillna(0)
test_data['phone_phoneFootprintStrengthOverall']=test_data['phone_phoneFootprintStrengthOverall'].fillna('None')
```

Feature Encoding


```python
#Finding categorical columns
categoical_cols=[]
for i in train_data:
    if train_data[i].dtype=='object':
        categoical_cols.append(i)
```

```python
#Label Encoding

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for i in categoical_cols:
    train_data[i] = label_encoder.fit_transform(train_data[i])
```

```python
train_data.head()
```

```python
#Same steps for test dataset
categoical_cols2=[]
for i in test_data:
    if test_data[i].dtype=='object':
        categoical_cols2.append(i)
```

```python
#Label Encoding

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for i in categoical_cols2:
    test_data[i] = label_encoder.fit_transform(test_data[i])
```

```python
test_data.head()
```

Feature Selection


```python
#Dropping irrelevant columns

X_train = train_data.drop(['Application Status'], axis=1)
y_train = train_data['Application Status']
```

```python
X_test = test_data.drop(['UID'], axis=1)
```

Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

#Splitting the training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Model Training
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

#Making predictions on the validation set
y_pred = model.predict(X_val)

#Model Evaluation
print("Accuracy:", accuracy_score(y_val, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_pred))
print("Classification Report:\n", classification_report(y_val, y_pred))
```

Test Predictions

```python
#Predicting on the test data
test_predictions = model.predict(X_test)

#Data Frame Creation
submission = pd.DataFrame({
    'UID': test_data['UID'],
    'Prediction': test_predictions
})

#Saving it as a csv file
submission.to_csv('predictions.csv', index=False)
```
